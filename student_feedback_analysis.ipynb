```python
import pandas as pd

df = pd.read_csv("student_feedback.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.fillna("", inplace=True)
```

```python
rating_cols = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

rating_cols
```

```python
for col in rating_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[rating_cols].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df[rating_cols].mean().plot(kind='bar')
plt.title("Average Rating for Each Question")
plt.xlabel("Feedback Question")
plt.ylabel("Average Score")
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df[rating_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Rating Questions")
plt.show()
```

```python
df["Total Score"] = df[rating_cols].sum(axis=1)
df["Average Score"] = df[rating_cols].mean(axis=1)
df.head()
```

```python
print("---- OVERALL SUMMARY ----")
print("Total Students:", len(df))
print("Overall Average Score:", df["Average Score"].mean())
print("\nAverage Score per Question:\n")
print(df[rating_cols].mean())
```
